```python
# Import dependencies
import pandas as pd
import numpy as np
```

```python
file_path = './resources/clean_steam_games.csv'
steam_df = pd.read_csv(file_path, index_col=0)
steam_df
```

```python
steam_df = steam_df.drop(columns=['all_reviews', 'original_price'])
steam_df
```

```python
steam_df.dtypes
```

```python
null_counts = steam_df['name'].isnull().sum()
print(null_counts)
```

```python
steam_df.dropna(subset=['name'], how='all', inplace=True)
```

```python
steam_df['popular_tags'].astype(str)
```

```python
tag_df = steam_df['popular_tags'].str.split(',', expand=True).rename(columns = lambda x: "tag_"+str(x+1))
tag_df
```

```python
tag_df['name'] = steam_df['name']
```

```python
col_name = 'name'
first_col = tag_df.pop(col_name)
tag_df.insert(0, col_name, first_col)
```

```python
tag_df.columns.tolist()
```

```python
merged_df = pd.merge(steam_df, tag_df, on='name')
merged_df
```

```python
merged_null_counts = merged_df['name'].isnull().sum()
print(merged_null_counts)
```

```python
genre_df = steam_df['genre'].str.split(',', expand=True).rename(columns = lambda x: "genre_"+str(x+1))
genre_df
```

```python
genre_df['name'] = steam_df['name']
```

```python
col_name = 'name'
first_col = genre_df.pop(col_name)
genre_df.insert(0, col_name, first_col)
```

```python
genre_df.columns.tolist()
```

```python
game_df = pd.merge(merged_df, genre_df, on='name')
game_df
```

```python
game_null_counts = game_df['name'].isnull().sum()
print(game_null_counts)
```

```python
game_df.to_csv('steam.csv', index=False)
```
