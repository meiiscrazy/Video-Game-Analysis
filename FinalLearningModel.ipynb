```python
# Import our dependencies
import pandas as pd
import matplotlib as plt
from sklearn.datasets import make_blobs
import sklearn as skl
import tensorflow as tf
```

```python
file_path="steamclean.csv"
Df=pd.read_csv(file_path)
Df.head(20)
```

```python
clean=Df.drop(columns=['genre_1','genre_2','genre_3','genre_4','genre_5','genre_6','genre_7','genre_8','genre_9','genre_10','genre_11','genre_12','genre_13','genre_14','genre_15','genre_16','average_rating'])
clean
```

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf
```

```python
#generating dummy columns for each genre and 
genre_dummies =clean['genres'].str.get_dummies(sep=';')
genre_dummies['name'] = clean['name']
col_name = 'name'
first_col = genre_dummies.pop(col_name)
genre_dummies.insert(0, col_name, first_col)
genremerged_df = pd.merge(clean, genre_dummies, on='name')
genremerged_df
```

```python
#performing the same for the categories columns and merging the categories and genres together
category_dummies =clean['categories'].str.get_dummies(sep=';')
category_dummies['name'] = clean['name']
col_name = 'name'
first_col = category_dummies.pop(col_name)
category_dummies.insert(0, col_name, first_col)
allmerged_df = pd.merge(genre_dummies, category_dummies, on='name')
allmerged_df
```

```python
#drop irrelevant columns here 
model_df=clean.drop(columns=['name','appid','release_date','developer','publisher','platforms','steamspy_tags', 'english','required_age','achievements','categories','genres'])
model_df
```

```python
model_cats = model_df.dtypes[model_df.dtypes == "object"].index.tolist()
model_cats
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(model_df[model_cats]))

# Add the encoded variable names to the DataFrame
encode_df.columns = enc.get_feature_names(model_cats)
encode_df.head()
```

```python
#merge the dummy datasets
final_df = pd.concat([allmerged_df, encode_df], axis=1)
final_df
```

```python
#drop irrelevant columns here 
ml_model=final_df.drop(columns=['name'])
ml_model
```

```python
#attempting to drop NaN values
ml_model.dropna()
```

```python
ml_model.columns.tolist()
```

```python
# Split our preprocessed data into our features and target arrays
y = ml_model["Free to Play"].values
X = ml_model.drop(["Free to Play", "Action", "owners_50000000-100000000"],1).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
len(X_train[0])
```

```python
# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net
number_input_features = len(X_train[0])
hidden_nodes_layer1= len(X_train[0])*3
hidden_nodes_layer2= 10

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="sigmoid"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train,y_train,epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
